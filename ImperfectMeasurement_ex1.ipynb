```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
# 1. MASUKKAN DATA PENGUKURAN DARI EXCEL-----------------------------------------  
df = pd.read_excel("DataArusTeganganIM.xlsx")
df.head()
```

```python
# Data x, nilai arus yang terukur pada resistor (mA)
x = df[['I (mA)']]
# Data y, tegangan yang terukur pada resistor (V)
y = df[['V (volt)']]
```

```python
#  2. PARAMETER REGRESI LINEAR----------------------------------------------
model = LinearRegression().fit(x,y)
tegangan_pred = model.predict(x)

print(model.coef_,model.intercept_)
```

```python
#  3a. PLOT GRAFIK-------------------------------------------
    #  plot data pengukuran-------------------------------------------
plt.scatter(x, y, label = 'data ukur')
plt.xlabel("Arus (mA)")
plt.ylabel("Tegangan (V)")

plt.legend()
plt.show()
```

```python
#  3b. PLOT GRAFIK-------------------------------------------
    # regresi linear dan plot--------------------------------------------
plt.scatter(x, y, label = 'data ukur')
plt.xlabel("Arus (mA)")
plt.ylabel("Tegangan (V)")
plt.plot(x,tegangan_pred, label="regresi", color = "red")
plt.title("Data pengukuran dengan $\hat{R}$ = " f"{model.coef_[0]:.3f} dan $b$ = {model.intercept_:.3f}")

plt.legend()
plt.show()
```
